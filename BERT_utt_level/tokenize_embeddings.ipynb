```python
import semEval as sem
import gensim
from gensim.models import Word2Vec
import csv
import re
```

```python
# read the json data into a dictionatry
data = sem.load_data()
# format the data so that all the relevant info is in a list or a list of lists
utt_emotions, cause_spans, utterances, speakers, emotions, dia_utt, all_vocab = sem.format_data(data)
```

# Tokenize data. 
Tokenizing the data takes awhile so read in the tokenized data instead of recalculating the same values

```python
# init the custom tokenizer
my_tokenizer = sem.Tokenizer()
# tokenize and padd the data. we only have to run this once since we have these values saved in tokenized_data
SPECIAL_TOKENS = ['<UNK>', '<PAD>', '<SOS>', '<EOS>']
tokenized_data = my_tokenizer(all_vocab)
vocab = sorted(set([w for ws in tokenized_data + [SPECIAL_TOKENS] for w in ws]))

with open('vocab.txt', 'w', encoding="utf-8") as vf:
    vf.write('\n'.join(vocab))
```

```python
# write tokenized data
with open('tokenized_data.csv', 'w', encoding = 'utf-8') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
    
    write.writerows(tokenized_data)
```

# READ IN TOKENS HERE!

```python
# read tokenized data
with open('tokenized_data.csv', 'r', encoding = 'utf-8') as read_obj: 
  
    # Return a reader object which will 
    # iterate over lines in the given csvfile 
    csv_reader = csv.reader(read_obj) 
  
    # convert string to list 
    tokenized_data = list(csv_reader) 
    # the reader is reading in empyt lists. we only wnat lists that are full
    tokenized_data = [sent for sent in tokenized_data if sent]
```

```python
# create embeddings using skipgram
# since skipgram randomly selects 'negative' semantic words, the word vectors will vary slightly
model2 = gensim.models.Word2Vec(tokenized_data, min_count = 1, vector_size = 100,
                                             window = 5, sg = 1)
word_vectors = model2.wv
```

```python
# play with the embeddings
print(model2.wv.most_similar('syphilis', topn = 10))
print(model2.wv.most_similar('Bitch', topn = 10))
print(model2.wv.most_similar('Boyfriend', topn = 10))
print(model2.wv.most_similar('Chandler', topn = 10))
print(model2.wv.most_similar('Friends', topn = 10))

model2.wv.similarity('teacher', 'syphilis')
```
